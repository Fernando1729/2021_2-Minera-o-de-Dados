```python
def state_name_conv(state_id):
    if state_id == 1:
        return 'New Delhi'
    elif state_id == 10:
        return 'Ahmedabad'
    elif state_id == 14:
        return 'Mumbai'
    elif state_id == 18:
        return 'Hyderabad'
    elif state_id == 19:
        return 'Bangalore'
    elif state_id == 20:
        return 'Chennai'
```

```python
Lista_1 = []


#['Toyota Rav4 \n',
 #'\n                            $14.380.000                        \n',
 #<ul class="card-list">
 #<li>Año: 2013</li>
 #<li>127.000 kms</li>
 #<li>Automática</li>
 #</ul>]


for i in Lista_Carros:

  item = i


  s  = str(item[0])
  # 'Toyota Rav4 \n'
  s1 = re.sub(' \n', '', s)   # 'Toyota Rav4'



  s  = str(item[1])
  #\n                            $14.380.000                        \n
  s  = re.sub('\n                            \$', '', s)
  #14.380.000                        \n
  s2 = re.sub('                        \n', '', s)    #14.380.000



  s  = str(item[2])
  # <ul class="card-list"><li>Año: 2013</li><li>127.000 kms</li><li>Automática</li></ul>]

  s  = re.sub('<ul class="card-list">\n<li>Año: ', '', s)
  # 2013</li><li>127.000 kms</li><li>Automática</li></ul>]
  
  s  = re.sub('</li>\n<li>', ',', s)
  # 2013,127.000 kms,Automática</li></ul>]
  
  s  = re.sub(' kms', '', s)
  # 2013,127.000,Automática</li></ul>]
  
  s3 = re.sub('</li>\n</ul>', '', s)        # 2013,127.000,Automática
  
  

  
  aux = s3.split(',', 2)

  #aux = str(x)

  s3 = aux[0]
  s4 = aux[1]
  s5 = aux[2]
  

  # [ 2013 , 127.000 , Automática ]


  Lista_1.append([s1, s2, s3, s4, s5 ])
  

print(Lista_1)

```
